```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    #Creates a normal distribution object dist with the calculated mean and standard deviation.
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
'''
get_pdf_probability takes in three parameters:

dataset: A list or array of numerical data.
startrange: The starting value of the range for which the probability is calculated.
endrange: The ending value of the range for which the probability is calculated.

sns.distplot(): Plots the distribution of the dataset. kde=True includes the kernel density estimate (KDE), 
which is a smoothed version of the histogram.
kde_kws={'color': 'blue'} sets the KDE line color to blue.
color='Green' sets the color of the histogram bars to green.
pyplot.axvline(): Draws vertical lines at startrange and endrange in red to indicate the specified range.

Assigns the dataset to sample.
Calculates the mean and standard deviation of the sample.
Prints the mean and standard deviation.

Creates a list of values from startrange to endrange.
Computes the PDF values for each value in the range using dist.pdf(value).
Sums up the PDF values to get the total probability for the specified range.
Prints the total probability (area under the curve) between startrange and endrange.

'''
```

```python
get_pdf_probability(dataset['ssc_p'],50,75)
```
