```python
dataset
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtypes=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual 
```

```python
quan,qual=quanQual(dataset)
```

```python
quan
```

```python
dataset[quan]
```

```python
'''
Why Use Imputers?
Maintain Data Integrity: Keeps the dataset intact without having to discard rows or columns with missing values.
Improve Model Performance: Many machine learning algorithms cannot handle missing values and imputing them allows these models 
to be trained effectively.
Reduce Bias: Proper imputation can help in reducing the bias that can be introduced by simply removing missing data.

'''
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
```

```python
imp.transform(dataset[quan])
```

```python
df=imp.transform(dataset[quan])
```

```python
df
```

```python
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
catedf=dataset[qual]
catedf
```

```python
frames=[df,catedf]
result=pd.concat(frames,axis=1)
display(result)
```
