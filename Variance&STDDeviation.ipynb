# IQR

```python
''' IQR Definition
In data science, the Interquartile Range (IQR) is a measure of statistical dispersion, which is the spread of the middle 50% of the data points.
It is a useful way to understand the variability within a dataset by focusing on the central portion, 
which can often be more informative than the overall range that can be skewed by outliers. The IQR is calculated as follows:

Sort the data: Arrange the data points in ascending order.

Find the first quartile (Q1): This is the median of the first half of the data. It represents the 25th percentile, 
meaning 25% of the data points are below this value.

Find the third quartile (Q3): This is the median of the second half of the data. It represents the 75th percentile, 
meaning 75% of the data points are below this value.

Calculate the IQR: Subtract Q1 from Q3.

IQR=Q3âˆ’Q1 '''
```

```python
descriptiveNew=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%",
                                "IQR","1.5rule","Lesser","Greater","MIN","MAX"],columns=quan)
for columnName in quan:
    descriptiveNew[columnName]["Mean"]=dataset[columnName].mean() # would not omit outliers
    descriptiveNew[columnName]["Median"]=dataset[columnName].median()# would omit outliers
    descriptiveNew[columnName]["Mode"]=dataset[columnName].mode()[0]# repeated quans
    descriptiveNew[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptiveNew[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptiveNew[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptiveNew[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptiveNew[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptiveNew[columnName]["IQR"]=descriptiveNew[columnName]["Q3:75%"]-descriptiveNew[columnName]["Q1:25%"]
    descriptiveNew[columnName]["1.5rule"]=1.5*descriptiveNew[columnName]["IQR"]
    descriptiveNew[columnName]["Lesser"]=descriptiveNew[columnName]["Q1:25%"]-descriptiveNew[columnName]["1.5rule"]
    descriptiveNew[columnName]["Greater"]=descriptiveNew[columnName]["Q3:75%"]+descriptiveNew[columnName]["1.5rule"]
    descriptiveNew[columnName]["MIN"]=dataset[columnName].min()
    descriptiveNew[columnName]["MAX"]=dataset[columnName].max()
    
```

```python
#Function creation
def freqTable(dataset,column):
    freqTableNew=pd.DataFrame(columns=["UniqueValues",'Frequency','RelativeFreq','CumulativeFreq'])
    freqTableNew['UniqueValues']=dataset[column].value_counts().index
    freqTableNew['Frequency']=dataset[column].value_counts().values
    freqTableNew['RelativeFreq']=(freqTableNew['Frequency']/freqTableNew.count()[0])
    freqTableNew['CumulativeFreq']=freqTableNew['RelativeFreq'].cumsum()
    return freqTableNew
```

# Variance & STD Deviation

```python
dataset['ssc_p'].var()
```

```python
'''
Variance
Variance measures the average squared deviation of each data point from the mean of the dataset. 
It gives an idea of the spread of the data points around the mean. 
A high variance indicates that the data points are spread out over a larger range of values, 
while a low variance indicates that the data points are closer to the mean.
'''
```

```python
dataset['ssc_p'].std()
```

```python
'''Standard Deviation
Standard deviation is the square root of the variance and provides a measure of the average distance of each data point from the mean.
It is expressed in the same units as the data, making it more interpretable in the context of the dataset.'''
```

```python
def univariate(dataset,quan):
    descriptiveNew=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%",
                                "IQR","1.5rule","Lesser","Greater","MIN","MAX","Kurtosis","Skewness",'Variance','STDDev'],columns=quan)
    for columnName in quan:
        descriptiveNew[columnName]["Mean"]=dataset[columnName].mean() # would not omit outliers
        descriptiveNew[columnName]["Median"]=dataset[columnName].median()# would omit outliers
        descriptiveNew[columnName]["Mode"]=dataset[columnName].mode()[0]# repeated quans
        descriptiveNew[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
        descriptiveNew[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
        descriptiveNew[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
        descriptiveNew[columnName]["99%"]=np.percentile(dataset[columnName],99)
        descriptiveNew[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
        descriptiveNew[columnName]["IQR"]=descriptiveNew[columnName]["Q3:75%"]-descriptiveNew[columnName]["Q1:25%"]
        descriptiveNew[columnName]["1.5rule"]=1.5*descriptiveNew[columnName]["IQR"]
        descriptiveNew[columnName]["Lesser"]=descriptiveNew[columnName]["Q1:25%"]-descriptiveNew[columnName]["1.5rule"]
        descriptiveNew[columnName]["Greater"]=descriptiveNew[columnName]["Q3:75%"]+descriptiveNew[columnName]["1.5rule"]
        descriptiveNew[columnName]["MIN"]=dataset[columnName].min()
        descriptiveNew[columnName]["MAX"]=dataset[columnName].max()
        descriptiveNew[columnName]["Kurtosis"]=dataset[columnName].kurtosis()
        descriptiveNew[columnName]["Skewness"]=dataset[columnName].skew()
        descriptiveNew[columnName]["Variance"]=dataset[columnName].var()
        descriptiveNew[columnName]["STDDev"]=dataset[columnName].std()
    return descriptiveNew
```
