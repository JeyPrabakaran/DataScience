```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
dataset = pd.read_csv('placement.csv')
dataset
```

```python
# Display the first few rows of the dataset to understand its structure
print(dataset.head())
```

```python
# Extract the 'ssc_p' column
ssc_p = dataset['ssc_p']
ssc_p
```

```python
# Calculate frequency
frequency, bins = np.histogram(ssc_p, bins=10)
'''The numpy.histogram function computes the histogram of a set of data.
#It takes an array as input and returns two arrays: one representing the counts (frequencies) of data points in each bin,
#and the other representing the bin edges.'''
'''ssc_p: This is the input array containing the data for which the histogram is to be computed.
bins=10: This specifies the number of bins to use for the histogram. In this case, the data will be divided into 10 equal-width bins.'''
```

```python
# Calculate relative frequency
relative_frequency = frequency / len(ssc_p)
```

```python
# Calculate cumulative frequency
cumulative_frequency = np.cumsum(relative_frequency)
```

```python
# Plot frequency histogram
plt.figure(figsize=(10, 6))
plt.hist(ssc_p, bins=10, color='blue', alpha=0.6)
#ssc_p: This is the input array containing the data for which the histogram is to be plotted.
#bins=10: This specifies the number of bins to use for the histogram. The data will be divided into 10 equal-width bins.
#color='blue': This sets the color of the bars in the histogram to blue.
#alpha=0.7: This sets the transparency level of the bars. The value ranges from 0 (completely transparent) to 1 (completely opaque).
#An alpha value of 0.7 means the bars will be slightly transparent.
plt.xlabel('SSC Percentage')
plt.ylabel('Frequency')
plt.title('Frequency Histogram of SSC Percentage')
plt.show()
```

```python
# Plot relative frequency histogram
plt.figure(figsize=(10, 6))
plt.hist(ssc_p, bins=10, weights=np.ones(len(ssc_p)) / len(ssc_p), color='green', alpha=0.7)
plt.xlabel('SSC Percentage')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Histogram of SSC Percentage')
plt.show()
```

```python
# Plot cumulative frequency histogram
plt.figure(figsize=(10, 6))
plt.hist(ssc_p, bins=10, weights=np.ones(len(ssc_p)) / len(ssc_p), cumulative=True, color='red', alpha=0.8)
plt.xlabel('SSC Percentage')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Frequency Histogram of SSC Percentage')
plt.show()
```
